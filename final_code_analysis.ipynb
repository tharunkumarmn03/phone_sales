```python
#check for null values

df.isnull().sum()
```

```python
#filling null values 

df['rating']=df['rating'].fillna(0)
```

```python
df['rating'].isnull().sum()
```

Adding 0 to null_values for the rating columns because the column is in numeric value and 0 will be treated as no review is given to the phones. 

```python
df['processor_speed'].isnull().sum()
```

```python
df['processor_speed']=df['processor_speed'].fillna('0')
```

```python
df['processor_speed']=df['processor_speed'].replace('Na','')
```

```python
df['processor_speed'].isnull().sum()
```

```python
df['fast_charging'].isnull().sum()
```

```python
df['fast_charging']=df['fast_charging'].fillna('0')
```

```python
# drop null values

df=df.dropna()
```

```python
df.isnull().sum()
```

```python
# describe() method returns description of the data in the DataFrame (i.e. count, mean, std, etc)

df.describe()
```


# Exploratory Data Analysis

```python
list(df.columns)
```

```python
# plotting a bar chart for brand_name and it's count

dff=df['brand_name'].value_counts().head(15)

dfff=dff.plot(kind ='bar',color='green')
for bars in dfff.containers:
    dfff.bar_label(bars)
sns.set(rc={'figure.figsize':(15,5)})
plt.title('List of brand')
```

       This is the top 15 highest sales acconding to the brand. Xiaomi is having a highest sales in the data 

```python
# plotting a bar chart for model vs price

phone_price=df[['model','price']].sort_values(by='price',ascending=False).head(10)

dff=sns.barplot(x='model',y='price',data=phone_price)
sns.set(rc={'figure.figsize':(10,5)})
plt.xticks(rotation=90)
plt.title ('Top_price')
for bars in dff.containers:
    dff.bar_label(bars)
```

  This is the top 10 highest phone price in the data. Vertu signature Touch is the higest phone price. 

```python
# plotting a bar chart for model vs price

phone_price=df[['model','price']].sort_values(by='price').head(12)

dff=sns.barplot(x='model',y='price',data=phone_price)
sns.set(rc={'figure.figsize':(10,5)})
plt.xticks(rotation=90)
plt.title ('Cheapest_phone')
for bars in dff.containers:
    dff.bar_label(bars)
```

This is the top 10 cheapest phone price in the data. Realme C2s is the cheapest phone price. 

```python
# plotting a bar chart for brand_name vs rating

brand_rating=df.groupby(['brand_name'],as_index=False)['rating'].sum().sort_values(by='rating',ascending=False).head(15)

sns.barplot(x='brand_name',y='rating',data=brand_rating)
sns.set(rc={'figure.figsize':(15,5)})
plt.title ('Company_rating')
```

From the above graph we can see that rating according to Brand given by the customer.Top rating according to barnds are Xiaomi , samsung ,vivo,  realme and oppo

```python
# plotting a bar chart for has_5g and it's count

dff=sns.countplot(x='has_5g', data =df)
sns.set(rc={'figure.figsize':(5,5)})
for bars in dff.containers:
    dff.bar_label(bars)
```

From the above bar graph we can see that more 5g mobile_phones are used.

```python
# plotting a bar chart for os and it's count

dff=sns.countplot(x='os', data=df)
for bars in dff.containers :
    dff.bar_label(bars)
sns.set(rc={'figure.figsize':(10,5)})    
```

From the above graphs Android OS is uesd in phone compared to others.

```python
# plotting a pie chart for fast_charging_available and it's percentage

dff=df['fast_charging_available'].value_counts()

dff.plot(kind='pie',autopct='%1.f%%')
sns.set(rc={'figure.figsize':(15,5)}) 
```

From the graphs you can see 87% fast_charging_avilable in the phones.

```python
# plotting a bar chart for brand_name vs rating

dff=df.groupby(['num_cores'],as_index=False)['processor_brand'].count()

dada=sns.barplot(x='num_cores',y='processor_brand', data=dff , hue='num_cores')
for bars in dada.containers:
    dada.bar_label(bars)
sns.set(rc={'figure.figsize':(8,5)}) 
```

From the above we can see that 8GB core is having more in the processer brand. 

```python
# plotting a bar chart for processor_brand and it's count

dff=sns.histplot(x='processor_brand',data =df )
sns.set(rc={'figure.figsize':(15,5)}) 
for bars in dff.containers :
    dff.bar_label(bars)
```

From the above graphs you see the Snapdragon processor is used more in the phone

```python
# plotting a bar chart for processor_brand vs os

dff=sns.countplot(x='processor_brand', data =df ,hue='os')
sns.set(rc={'figure.figsize':(15,5)}) 
for bars in dff.containers:
    dff.bar_label(bars)
```

 From the above graphs we can see more snapdragon processor is used in phone by comparing with the os.

```python
# plotting a hist chart for processor_speed and it's count

sns.histplot(x='processor_speed', data=df, bins=10)
sns.set(rc={'figure.figsize':(10,5)})
```

From the above graphs we can see that from the range of 1.5 to 2.5 have the higest count of processor_speed in the mobiles.
